## Query

```python
df[ #Isi query atau filter
    (df['year_id']==2001) &
    (df['fran_id']=='Celtics') &
    (df['game_location']=='H') &
    (df['game_result']=='W')   
]['opp_fran']
```

```python
df[ #Isi query atau filter
    (df['year_id']==2001) &
    (df['fran_id']=='Celtics') &
    (df['game_location']=='H') &
    (df['game_result']=='W')   
][['fran_id','pts','opp_fran','opp_pts']] #Ingat pakai [[]] karena kolom yang ditarik ada di dalam list
```

Alternative

```python
df.query(
    "year_id==2001 &\
    fran_id=='Celtics' &\
    game_location=='H' &\
    game_result=='W'"
)['opp_fran']
```

## Grouping

```python
df.groupby('game_result')['pts'].sum()
```

```python
df.groupby('game_result')['pts'].mean()
```

Query + Group By

```python
df[
    (df['year_id']>=2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','game_result'])['pts'].mean()
```

Aggregate

```python
df[
    (df['year_id']>=2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','game_result'])['pts'].agg(['min','max'])
```

```python
df[
    (df['year_id']>=2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','game_result']).agg({'pts':'mean','game_id':'count'})
```
