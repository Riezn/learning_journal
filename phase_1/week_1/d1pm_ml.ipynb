## 5. Data Processing

## Data Inference

```python
data_inf = df.sample(10, random_state=11)
data_inf
```

```python
df_tt = df.drop(data_inf.index)
df_tt
```

```python
df_tt.reset_index(drop=True,inplace=True)
data_inf.reset_index(drop=True,inplace=True)
```

## Splitting Dataset

```python
X = df_tt.drop(['Rating'],axis=1)
y = df_tt['Rating']
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

print(X_train.shape)
print(X_test.shape)
```

## Feature Selection

```python
X_train
```

**Asumsi**

`Name` tidak ada hubungannya dengan `Rating`, maka akan di-drop. Sebisa mungkin pakai sumber literatur untuk drop kolom.

```python
X_train.drop(columns='Name',inplace=True)
X_test.drop(columns='Name',inplace=True)
```

Bagi dua masing-masing train dan test set berdasarkan data (kolom)-nya:
- Numeric -> Feature Scaling
- Categoric -> Feature Encoding

Gabung kembali menjadi `train-set` -> `model_train` dan `test-set` -> `predict`

```python
num_columns = X_train.select_dtypes(include=np.number).columns.tolist()
cat_columns = X_train.select_dtypes(include=['object']).columns.tolist()

print('Nume Cols: ', num_columns)
print('Cat Cols: ', cat_columns)
```

```python
X_train_num = X_train[num_columns]
X_train_cat = X_train[cat_columns]

X_test_num = X_test[num_columns]
X_test_cat = X_test[cat_columns]
```

## Feature Scaling

```python
X_train_num
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train_num)

X_train_num_scl = scaler.transform(X_train_num)
X_test_num_scl = scaler.transform(X_test_num)

X_train_num_scl
```

## Feature Encoding

```python
X_train_cat
```

```python
from sklearn.preprocessing import OrdinalEncoder #OrdinalEncoder untuk Ordinal
encoder = OrdinalEncoder()
encoder.fit(X_train_cat)

X_train_cat_enc = encoder.transform(X_train_cat)
X_test_cat_enc = encoder.transform(X_test_cat)

X_train_cat_enc
```

## Concat Num and Cat

```python
X_train_final = np.concatenate([X_train_num_scl,X_train_cat_enc], axis=1)
X_test_final = np.concatenate([X_test_num_scl,X_test_cat_enc], axis=1)

X_train_final
```

```python
num_columns
```

```python
X_train_final_df = pd.DataFrame(X_train_final, columns=[num_columns + cat_columns])
X_train_final_df
```

## 6. Model Definition

```python
from sklearn.linear_model import LinearRegression
model_lin_reg = LinearRegression()
```

Metrics yang akan dipakai adalah `MAE` (Mean Absolute Error)

## 7. Model Training

```python
model_lin_reg.fit(X_train_final,y_train)
```

## 8. Model Evaluation

```python
y_pred_train = model_lin_reg.predict(X_train_final)
y_pred_test = model_lin_reg.predict(X_test_final)

y_pred_test
```

```python
from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)

print('Error Train: ', mae_train)
print('Error Test: ', mae_test)
```
