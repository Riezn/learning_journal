```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fahmimnalfrzki/Dataset/main/kc_house_data.csv')
data
```

```python
df = data.copy()
```

```python
df.columns
```

```python
data_inf = df.sample(10, random_state=11)
data_inf
```

```python
df_tt = df.drop(data_inf.index)
df_tt
```

```python
df_tt.reset_index(drop=True,inplace=True)
data_inf.reset_index(drop=True,inplace=True)
```

```python
X = df_tt.drop(columns=(['id','date','price']))
y = df_tt['price']
```

```python
X 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train_scaled,y_train)

y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test_scaled)
```

```python
import sklearn.metrics as skm

mae_train = skm.mean_absolute_error(y_train, y_pred_train)
r2_train = skm.r2_score(y_train, y_pred_train)
mape_train = skm.mean_absolute_percentage_error(y_train, y_pred_train)

mae_test = skm.mean_absolute_error(y_test, y_pred_test)
r2_test = skm.r2_score(y_test, y_pred_test)
mape_test = skm.mean_absolute_percentage_error(y_test, y_pred_test)

print('TRAIN')
print('MAE: ', mae_train)
print('MAPE: ', mape_train)
print('R2: ', r2_train)

print('TEST')
print('MAE: ', mae_test)
print('MAPE: ', mape_test)
print('R2: ', r2_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline


for deg in np.arange(1,5,1):
    poly_model = make_pipeline(PolynomialFeatures(deg),LinearRegression())
    poly_model.fit_transform(X_train_scaled)
    
    y_poly_pred_train = poly_model.predict(X_train_scaled)
    y_poly_pred_test = poly_model.predict(X_test_scaled)

    print('DEG: ', deg)
    print('TRAIN')
    print('MAE: ', mae_train)
    print('MAPE: ', mape_train)
    print('R2: ', r2_train)

    print('TEST')
    print('MAE: ', mae_test)
    print('MAPE: ', mape_test)
    print('R2: ', r2_test)
```
