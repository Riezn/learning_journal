```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fahmimnalfrzki/Dataset/main/kc_house_data.csv')
data
```

```python
df = data.copy()
```

```python
data_inf = df.sample(10, random_state=11)
data_inf
```

```python
df_tt = df.drop(data_inf.index)
df_tt
```

```python
df_tt.reset_index(drop=True,inplace=True)
data_inf.reset_index(drop=True,inplace=True)
```

```python
df_tt.columns
```

```python
X = df_tt.drop(columns=(['id','date','waterfront']))
y = df_tt['waterfront']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(penalty='l2',solver='newton-cg',C=0.1,fit_intercept=True)
```

```python
model.fit(X_train_scaled,y_train)
```

```python
y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test_scaled)
```

```python
print('Train Accuracy: ',accuracy_score(y_train,y_pred_train))
print('Train Accuracy: ',accuracy_score(y_test,y_pred_test))
```
